### **Question - ATM Dispatch**
* Only Rs. 500 and Rs. 200 notes are available.
* Take an amount as input from user
  * if amount is dischargeable, `print("Discharging cash")`
  * else, `print("Denomination not available")`

````python
amount = int(input('Enter amount: '))
if (amount%500)%200 == 0 :
  print('Discharging cash')
else:
  print('Denomination not available')

# Enter 600
````

- But there's a problem with this solution.
- This code checks if the remainder amount after dispensing Rs. 500 notes can be dispensed in Rs. 200 notes.
- But if we enter an amount which can be dispensed using only Rs. 200 notes (e.g. 600), it fails to give the desired output.

````python
amount=int(input('Enter amount: '))
if amount%500==0 or amount%200==0:
  print("Discharging cash")
else:
  print("Denomination not available")

# Enter 1700
````

- But there's a problem with this solution as well.
- This code checks whether the entered amount can be dispensed using either Rs. 500 notes or Rs. 200 notes.
- But if we enter an amount which can be dispensed using a combination of both the notes (e.g. 1700), it fails to give the desired output.

We will see how to solve the problems in these approches in a few minutes.

### **Solution for the ATM Dispatch problem -**

````python
amount = int(input('Enter amount: '))
if amount <= 500 and (amount%500)%200 == 0:
  print('Discharging cash')
elif amount > 500 and ((amount%500)%200)%100 == 0:
  print('Discharging cash')
else:
  print('Denomination not available')
````
