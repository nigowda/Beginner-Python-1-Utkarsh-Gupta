### **What is a Tuple?**

* A tuple is sort of like a immutable list.

Screenshot 2022-12-06 101852.png

* In case of a list, we can change values of the elements.

````python
a = [1, 2, 3, 4, 5]
a[0] = 10
a
````

* In case of a tuple, we cannot change the values.

````python
a = (1, 2, 3, 4, 5)
a[0] = 10
a
````

### **How to create a tuple?**

* The notation to create tuples is same as lists but instead of **square brackets** `[]` we use **parentheses** `()`.

Screenshot 2022-12-06 102528.png

* We can do **positive & negative indexing** on tuples.
* We can even **slice** tuples.

````python
a = (1, 2, 3, 4, 5, 6, 7, 8, 9, 0)
````

````python
a[2:8]
````

````python
a[-1:-8]
````

### **Defining tuples with a single element -**

````python
# This will be a tuple
t = ()
type(t)
````

````python
# This will be a tuple
t = (1, 2, 3, 4)
type(t)
````

````python
# This will not be a tuple
t = (45)
type(t)
````

**To define a single element tuple, we add a comma at the end of that element.**

````python
t = (45,)
type(t)
````

````python
t = 45, # without parentheses - also a tuple
type(t)
````

### **Type casting tuples -**

````python
a = [1, 2, 3, 4, 5]
b = tuple(a)
print(a) # List
print(b) # Tuple
````

### **Python uses tuples in its internal implementation -**

````python
def foo():
    return 200, 300, 400

bar = foo()
print(bar) # Tuple
````

### **Iterating over a list of tuples -**

````python
a = [(1, "Bipin"), (2, "Ranjan"), (3, "Shanoor"), (4, "Shital"), (5, "Tharoor"), ]
````

````python
# Prints all the tuples -
for i in a:
    print(i)
````

````python
# We can also do something like -
for roll_no, name in a:
    print(f"Roll Number - {roll_no} is {name}!")
````

### **Sets**

* A Set is a collection of unique elements.
* A Set is represented using curly braces `{ }`.

Screenshot 2022-12-06 105340.png

````python
s = {1, 2, 3, 4, 5, 5, 5, 5, 4, 4, 4, 7, 7, 7}
s
````

* Sets in Python are not ordered.

````python
# Error, because there is no order.
s[0]
````

* To create an empty set, we have to use the `set()` function.

````python
s = set()
s
````

**Note:**
* Sets can store `strings`, `tuples`, and `booleans` inside them all at once.
* Sets cannot store `lists`, `sets`, and `dictionaries` within them.

## **Set Methods**

* There are very limited number of methods for a Set.

````python
s = {1, 2, 3}
s
````

#### `.add()` - Add any element at the end of the Set.

````python
s.add(4)
s
````

#### `.remove()` - Remove any element from the Set.

````python
s.remove(3)
s
````

#### `.pop()` - Remove the first element from the Set.

````python
s.pop()
s
````

#### `.update()` - Update the elements in the Set using a list/tuple of elements.

````python
s.update([4, 5, 6, 6, 7, 9, 1])
s
````

````python
s.update((3, 0, 0, 8))
s
````

### **Set Operations**
* There are many mathematical operations we can perform on Sets.

#### **Intersection**
* An intersection of two sets contains the common elements in both sets.

Screenshot 2022-12-06 111332.png

````python
s1 = {2, 3, 4, 5}
s2 = {1, 3, 4, 6, 7, 8}
````

````python
s1.intersection(s2)
````

````python
s1 & s2
````

#### **Union**
* A union of two sets contains all the elements from both sets.

Screenshot 2022-12-06 111341.png

````python
s1 = {2, 3, 4, 5}
s2 = {1, 3, 4, 6, 7, 8}
````

````python
s1.union(s2)
````

````python
s1 | s2
````

#### **Difference**
* The difference of two sets - A and B, contains all elements in **A** which are not present in **B**.

Screenshot 2022-12-06 111350.png

````python
s1 = {2, 3, 4, 5}
s2 = {1, 3, 4, 6, 7, 8}
````

````python
s1.difference(s2)
````

````python
s1 - s2
````

#### **Symmetric Difference**
* The symmetric difference of two sets contains all the elements in **union** except the **common elements**.

Screenshot 2022-12-06 111405.png

````python
s1 = {2, 3, 4, 5}
s2 = {1, 3, 4, 6, 7, 8}
````

````python
s1.symmetric_difference(s2)
````

````python
s1 ^ s2
````

### **Question - 3**

* You are given a string as input.
* Ascertain whether that string contains a binary number or not.

```
Sample input 1
"0011010101ab"
```

```
Sample output 1
"Not Binary!"
```

```
Sample input 2
"010010101010"
```

```
Sample output 2
"Binary!"
```

````python
# Method 1
def check_binary(string):
    for i in string:
        if not (i == "0" or i == "1"):
            return "Not Binary!"
    return "Binary!"
````

````python
print(check_binary("adasfafasfas"))
print(check_binary("01010101010"))
print(check_binary("11111"))
````

````python
# Method 2
def check_binary_cool(string):
    s = set(string)
    if s <= {"0", "1"}:
        return "Binary!"
    return "Not Binary!"
````

````python
print(check_binary_cool("adasfafasfas"))
print(check_binary_cool("01010101010"))
print(check_binary_cool("11111"))
````
